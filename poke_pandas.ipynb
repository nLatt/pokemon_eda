```python
import pandas as pd
import json

with open("pokedex.json") as pokedex_json:
    pokedex_str = pokedex_json.read()
pokedex = json.loads(pokedex_str)["pokemon"]

pokemon_df = pd.DataFrame(pokedex) 
pokemon_df.head()
len(pokemon_df)
```

```python
try:
    pokemon_df["weight"] = pokemon_df.apply(lambda pok: pok["weight"].replace(" kg", ""), axis=1)
    pokemon_df["weight"] = pokemon_df.apply(lambda pok: float(pok["weight"]), axis=1)
except:
    pass

pokemon_df["candy_count"].fillna(value=0, inplace=True)
pokemon_df["next_evolution"] = pokemon_df["next_evolution"].apply(lambda x: 0 if type(x) != list else x) 
pokemon_df["prev_evolution"] = pokemon_df["prev_evolution"].apply(lambda x: 0 if type(x) != list else x) 
pokemon_df.head()
```

```python
# DF with the num, name and weight of all pokemon that weigh higher than 10 kg

pokemon_weight = pokemon_df.loc[pokemon_df["weight"] > 10]
pokemon_weight = pokemon_weight[["num", "name", "weight"]].sort_values(by=["weight"])
pokemon_weight
```

```python
# DF with the pokemon and their evolutions

pokemon_evo = pokemon_df.loc[pokemon_df["prev_evolution"] == 0]
pokemon_evo = pokemon_evo[["id", "name", "next_evolution"]]

pokemon_evo["second_evo"] = pokemon_evo["next_evolution"].apply(lambda x: x[0]["name"] if type(x) == list else 0)
pokemon_evo["third_evo"] = pokemon_evo["next_evolution"].apply(lambda x: x[1]["name"] if type(x) == list and len(x) == 2 else 0)
pokemon_evo = pokemon_evo.drop(["next_evolution"], axis=1)

pokemon_evo
```
